```python
import pandas as pd
dataset = "/content/drive/MyDrive/Colab Notebooks/MultilingualDataset-classification-NLP-HW4 - Multilingual.csv"
```

```python
# Load data
data = pd.read_csv(dataset)
data.dropna()
print(data.head(3))
```

```python
english_true_label = data['TRUE LABEL']
english_chatgpt_label = data['CHATGPT']
english_llama_label = data['LLAMA']
arabic_true_label = data['arabic TRUE LABEL']
arabic_chatgpt_english_prompt = data['arabic CHATGPT - English prompt']
arabic_chatgpt_arabic_prompt = data['arabic CHATGPT - Arabic prompt']
arabic_llama_english_prompt = data['arabic LLAMA- English Prompt']
arabic_llama_arabic_prompt = data['arabic LLAMA- Arabic Prompt']
french_true_label = data['french TRUE LABEL']
french_chatgpt_english_prompt = data['french CHATGPT- Englsh Prompt']
french_chatgpt_french_prompt = data['french CHATGPT-French Prompt']
french_llama_english_prompt = data['french LLAMA - English prompt']
french_llama_french_prompt = data['french LLAMA - French Prompt']
spanish_true_label = data['spanish TRUE LABEL']
spanish_chatgpt_english_prompt = data['spanish CHATGPT- English prompt']
spanish_chatgpt_spanish_prompt = data['spanish CHATGPT-Spanish Prompt']
spanish_llama_english_prompt = data['spanish LLAMA - English Prompt']
spanish_llama_spanish_prompt = data['spanish LLAMA - Spanish Prompt']
italian_true_label = data['italian TRUE LABELS']
italian_chatgpt_english_prompt = data['italian CHATGPT- English Prompt']
italian_chatgpt_italian_prompt = data['italian CHATGPT-Italian Prompt']
italian_llama_english_prompt = data['italian LLAMA -  English Prompt']
italian_llama_italian_prompt = data['italian LLAMA - Italian Prompt']
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

def printClassificationReport(true_labels, predict_labels):
  report = classification_report(true_labels, predict_labels)
  confusionMatrix = confusion_matrix(true_labels, predict_labels)
  return(report, confusionMatrix)
```

```python
report, confusionMatrix = printClassificationReport(english_true_label, english_chatgpt_label)
print("\nCHATGPT- ENGLISH\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(english_true_label, english_llama_label)
print("\nLLAMA-ENGLISH\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(arabic_true_label, arabic_chatgpt_english_prompt)
print("\nCHATGPT- ARABIC - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(arabic_true_label, arabic_chatgpt_arabic_prompt)
print("\nCHATGPT- ARABIC - Arabic prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(arabic_true_label, arabic_llama_english_prompt)
print("\nLLAMA- ARABIC - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(arabic_true_label, arabic_llama_arabic_prompt)
print("\nLLAMA- ARABIC - Arabic prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(french_true_label, french_chatgpt_english_prompt)
print("\nCHATGPT- FRENCH - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(french_true_label, french_chatgpt_french_prompt)
print("\nCHATGPT- FRENCH - French prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(french_true_label, french_llama_english_prompt)
print("\nLLAMA- FRENCH - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(french_true_label, french_llama_french_prompt)
print("\nLLAMA- FRENCH - French prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(spanish_true_label, spanish_chatgpt_english_prompt)
print("\nCHATGPT- SPANISH - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(spanish_true_label, spanish_chatgpt_spanish_prompt)
print("\nCHATGPT- SPANISH - Spanish prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(spanish_true_label, spanish_llama_english_prompt)
print("\nLLAMA- SPANISH - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(spanish_true_label, spanish_llama_spanish_prompt)
print("\nLLAMA- SPANISH - Spanish prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(italian_true_label, italian_chatgpt_english_prompt)
print("\nCHATGPT- ITALIAN - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(italian_true_label, italian_chatgpt_italian_prompt)
print("\nCHATGPT- ITALIAN - Italian prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(italian_true_label, italian_llama_english_prompt)
print("\nLLAMA- ITALIAN - English prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
report, confusionMatrix = printClassificationReport(italian_true_label, italian_llama_italian_prompt)
print("\nLLAMA- ITALIAN - Italian prompt\n", report,"\nConfusion Matrix : \n",confusionMatrix)
```
